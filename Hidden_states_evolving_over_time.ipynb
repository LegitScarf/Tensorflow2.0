```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, SimpleRNN
```

```python
T = 20
X = np.sin(np.linspace(0, 4 * np.pi, T))
X = X.reshape(1, T, 1)
```

```python
i = Input(shape=(T, 1))
x = SimpleRNN(5, activation='tanh', return_sequences=True)(i)
model = Model(i, x)

hidden_states = model.predict(X)[0]
```

```python
plt.figure(figsize=(10, 5))
for unit in range(5):
    plt.plot(hidden_states[:, unit], label=f'Hidden Unit {unit+1}')

plt.xlabel("Time Step")
plt.ylabel("Hidden State Value")
plt.title("Evolution of Hidden States Over Time")
plt.legend()
plt.grid()
plt.show()
```
